Create a new column 'patched_values' which contains the same values as the 'vals' any negative values in 'vals' with the group mean:

```
    vals grps  patched_vals
0    -12    A          13.6
1     -7    B          28.0

3      4    A           4.0

8      8    A           8.0
```

````python
import pandas as pd
import numpy as np
df = pd.DataFrame({"vals": np.random.RandomState(31).randint(-30, 30, size=15), 
                   "grps": np.random.RandomState(31).choice(["A", "B"], 15)})
df[(df['vals']>0) & (df['grps']=='A')].vals.mean()
df['patched_vals'] = np.where(df['vals']>=0, df['vals'], np.where((df['vals']<0) & (df['grps']=="A"), df[(df['vals']>0) & (df['grps']=='A')].vals.mean(), df[(df['vals']>0) & (df['grps']=='B')].vals.mean()))
#here I use nested np.where to meet multiple condition. since if it was a normal code then I needed to us if, elif and else condition.
df
````
